```python
import pandas as pd

# Llegeix el fitxer CSV
df = pd.read_csv('dades.csv')

# 1. **La mitjana d’edat dels individus**
mitjana_edat = df['Edat'].mean()
print(f"La mitjana d'edat de tots els individus és: {mitjana_edat:.2f} anys")

# 2. **La distribució de gèneres**
distribucio_genere = df['Gènere'].value_counts(normalize=True) * 100
print("\nDistribució de gèneres en percentatges:")
print(distribucio_genere)

# 3. **Ciutats amb més representació**
ciutats_mes_representades = df['Ciutat'].value_counts()
print("\nCiutats més representades:")
print(ciutats_mes_representades)

# 4. **Persones amb l'edat mínima i màxima**
edat_minima = df['Edat'].min()
edat_maxima = df['Edat'].max()
persones_edat_minima = df[df['Edat'] == edat_minima]
persones_edat_maxima = df[df['Edat'] == edat_maxima]

print("\nPersones amb l'edat mínima:")
print(persones_edat_minima)

print("\nPersones amb l'edat màxima:")
print(persones_edat_maxima)

# 5. **Distribució per professions**
distribucio_professio = df['Professió'].value_counts()
print("\nDistribució de professions:")
print(distribucio_professio)
```
