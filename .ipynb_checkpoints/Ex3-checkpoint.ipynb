3. ReportLab
Funcionalitat principal:

ReportLab és una llibreria de Python que permet generar documents PDF programàticament.
S'utilitza per crear informes, factures, i documents personalitzats amb formats professionals.

Funcions clau:
Crear PDFs des de zero, incloent-hi text, taules, imatges, i gràfics.
Personalitzar el disseny i els estils dels documents.
Generació automàtica d'informes basats en dades.

```python
import pandas as pd

# Crear un DataFrame a partir d'un diccionari
data = {
    'Nom': ['Anna', 'Joan', 'Maria'],
    'Edat': [23, 35, 29],
    'Ciutat': ['Barcelona', 'València', 'Girona']
}
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)

# Filtrar persones majors de 30 anys
majors_30 = df[df['Edat'] > 30]
print(majors_30)

# Calcular l’edat mitjana
edat_mitjana = df['Edat'].mean()
print(f"Edat mitjana: {edat_mitjana}")
```

```python
# Celdas en Jupyter Notebook

# Celda 1: Importar librerías
import numpy as np
import matplotlib.pyplot as plt

# Celda 2: Crear datos
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Celda 3: Visualizar los datos
plt.plot(x, y)
plt.title("Gráfico de una onda")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
from reportlab.pdfgen import canvas

# Crear un lienzo para el PDF
pdf = canvas.Canvas("ejemplo.pdf")

# Agregar texto
pdf.drawString(100, 750, "¡Hola, este es un PDF generado con ReportLab!")
pdf.drawString(100, 730, "Puedes agregar texto, gráficos, e imágenes.")

# Dibujar una línea
pdf.line(100, 720, 400, 720)

# Guardar el archivo PDF
pdf.save()

print("El archivo PDF 'ejemplo.pdf' ha sido generado.")
```
