De la mateixa manera que a l'exercici 5 caldrà investigar una llibreria per a realitzar gràfics de dades visuals i elaborar com a mínim el codi necessari per presentar de forma que tinguin sentit amb l'àmbit del problema.
Com a restricció caldrà com a mínim generar 4 gràfiques coherents.
Es pot utilitzar la llibreria python matploit

```python
import pandas as pd
import matplotlib.pyplot as plt

# Llegeix el fitxer CSV
df = pd.read_csv('dades.csv')

# 1. **Mitjana d'edat**
mitjana_edat = df['Edat'].mean()
print(f"La mitjana d'edat de tots els individus és: {mitjana_edat:.2f} anys")

# 2. **Distribució de gèneres**
distribucio_genere = df['Gènere'].value_counts()

# Gràfica 1: Distribució de gèneres
plt.figure(figsize=(6, 6))
distribucio_genere.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'pink'])
plt.title("Distribució de gèneres")
plt.ylabel("")  # Elimina etiqueta vertical
plt.show()

# 3. **Ciutats amb més representació**
ciutats_mes_representades = df['Ciutat'].value_counts()

# Gràfica 2: Ciutats més representades
plt.figure(figsize=(10, 6))
ciutats_mes_representades.plot(kind='bar', color='lightgreen')
plt.title("Ciutats més representades")
plt.xlabel("Ciutats")
plt.ylabel("Nombre de persones")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 4. **Edat mínima i màxima**
edat_minima = df['Edat'].min()
edat_maxima = df['Edat'].max()
persones_edat_minima = df[df['Edat'] == edat_minima]
persones_edat_maxima = df[df['Edat'] == edat_maxima]

# Gràfica 3: Histograma d'edats
plt.figure(figsize=(8, 6))
df['Edat'].plot(kind='hist', bins=8, color='orange', edgecolor='black')
plt.title("Distribució de les edats")
plt.xlabel("Edat")
plt.ylabel("Nombre de persones")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 5. **Distribució per professions**
distribucio_professio = df['Professió'].value_counts()

# Gràfica 4: Distribució de professions
plt.figure(figsize=(10, 6))
distribucio_professio.plot(kind='barh', color='mediumpurple')
plt.title("Distribució de professions")
plt.xlabel("Nombre de persones")
plt.ylabel("Professió")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
